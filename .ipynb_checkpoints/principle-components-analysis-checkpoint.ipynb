# Principle components analysis (PCA) implemented in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets.samples_generator import make_blobs
from sklearn.decomposition import PCA
```

## Dataset

```python
X, y = make_blobs(n_samples = 10000, n_features = 3, centers = [[3, 3, 3], [0, 0, 0], [1, 1, 1], [2, 2, 2]],
                 cluster_std = [0.2, 0.1, 0.1, 0.2], random_state = 9)
fig = plt.figure(figsize = (12, 8))
ax = Axes3D(fig, rect = [0, 0, 1, 1], elev = 30, azim = 20)
plt.scatter(X[:, 0], X[:, 1], X[:, 2], marker = 'o', c = y)
```

## PCA process

```python
pca = PCA(n_components = 3)
pca.fit(X)
print('pca.explained_variance_ratio_: ', pca.explained_variance_ratio_)
print('pca.explained_variance_: ', pca.explained_variance_)
```

```python
pca = PCA(n_components = 2)
pca.fit(X)
print('pca.explained_variance_ratio_: ', pca.explained_variance_ratio_)
print('pca.explained_variance_: ', pca.explained_variance_)
```

```python
X_new = pca.transform(X)
plt.scatter(X_new[:, 0], X_new[:, 1], marker = 'o', c = y)
plt.show()
```

```python
pca = PCA(n_components = 0.95)
pca.fit(X)
print('pca.explained_variance_ratio_: ', pca.explained_variance_ratio_)
print('pca.explained_variance_: ', pca.explained_variance_)
print('pca.n_components_: ', pca.n_components_)
```
